```python
colours = ["red", "green", "blue"]
```

## ``for`` loop

```python
for colour in colours:
    print(colour)
```

```python
shouted_colours = []

# for each colour in the list called colours
for colour in colours:
    # make it uppercase and add exclamation points
    shouted_colour = colour.upper()+"!!" 
    # add this to the external list
    shouted_colours.append(shouted_colour)
    # print to screen
    print(shouted_colour)
```

```python
shouted_colours
```

# ``if/else``

```python
for colour in colours:
    if colour == "green":
        print(colour.upper()+"!!")
    elif colour == "red":
        print("*"+colour+"*")
    else:
        print(colour)
```

```python
weird_list = ["red", "green", 10, "blue"]
```

```python
# an if/else statement that implements exceptions 
for value in weird_list:
    if type(value) == int:
        print(value/2)

    else:
        print("Not a number!")
```

```python
# another way to implement exceptions in an if/else statement
for value in weird_list:
    if type(value) == int:
        print(value/2)

    else:
        pass 
```

### Exception handling

```python
for value in weird_list:
    try: 
        print(value/2)
    except: # if you run into an exception/error, here you don't have to type out the specific exception(s)
        print("Not a number!") 
```

### Document length

```python
tokens = text.lower().split() # making everything lowercase and splitting by whitespaces
```

```python
tokens
```

### Counting word frequencies

```python
keyword = "love"
```

```python
tokens.count(keyword) 
```

```python
import string
```

```python
string.punctuation # the punctuations we will encounter when working with strings in python
```

```python
# List comprehensions
cleaned_tokens = [''.join(char for char in tokens if char not in string.punctuation) for token in tokens] 
```
